# Difference between trials

This notebook examines the difference in the distribution of EEGNet's SoftMax predictions between episodes the copilot succesfully acquires the target and episodes were the copilot fails to acquire the target. 

```python
from TaskData import TaskData, plot_softmax_distribution
import matplotlib.pyplot as plt
```

# Progressive training experiments

In this first set of experiments, a pretrained copilot is succesively fine tuned on 10-20 minute data sessions for a total of around 100 minutes of data. The distribution of SoftMax probabilities accross trials where the copilot succesfully hits the target (left) and misses the target (right) are shown for each of these datasets.

This copilot was furhter trained using a larger shaped reward (Fs = 30,000) and a single gradient update per timestep. Overall, this copilot/agent did not succesfully adopt a meaningful policy after 100 minutes of experience with replayed human data.

```python
data = TaskData("/home/brandon/eeg/bci_raspy/data/part11")
data.partition_trials()
data.compute_accuracy()

for ix in range(4):
    plt.figure(figsize=(16,6))
    plt.subplot(121)
    plt.title("distribution of SoftMax on Correct Trials")
    left_trial_probs = data.get_softmax_probs(trial_type=ix, all_trials='correct')
    plot_softmax_distribution(left_trial_probs, bins=6)
    plt.legend(["Left", "Right", "Up", "Down"])

    plt.subplot(122)
    plt.title("distribution of SoftMax on Incorrect Trials")
    left_trial_probs = data.get_softmax_probs(trial_type=ix, all_trials='miss')
    plot_softmax_distribution(left_trial_probs, bins=6)
    plt.legend(["Left", "Right", "Up", "Down"])
plt.show()
```

## Retested NFQ with bug fix

Here NFQ was used to train for 1,000 steps and BM_4_movement_w_feedback_111 was replayed to generate the following results. There isn't much of a performance benefit of the NFQ but it also hasn't hurt the pre-existing performance of the pretrained copilot. Next steps should be to focus on improving the NFQ training.
